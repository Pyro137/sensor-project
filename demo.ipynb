```python
import numpy as np

print("df1['ac_000'] info:")
print(df1["ac_000"].describe())
print(f"NaN count: {df1['ac_000'].isna().sum()}")
print(f"Infinite count: {np.isinf(df1['ac_000']).sum()}")

print("\ndf2['ac_000'] info:")
print(df2["ac_000"].describe())
print(f"NaN count: {df2['ac_000'].isna().sum()}")
print(f"Infinite count: {np.isinf(df2['ac_000']).sum()}")

# Remove NaN and infinite values before performing the test
df1_ac_000_clean = df1["ac_000"].dropna().replace([np.inf, -np.inf], np.nan).dropna()
df2_ac_000_clean = df2["ac_000"].dropna().replace([np.inf, -np.inf], np.nan).dropna()

print("\nAfter cleaning:")
print(f"df1['ac_000'] length: {len(df1_ac_000_clean)}")
print(f"df2['ac_000'] length: {len(df2_ac_000_clean)}")

if len(df1_ac_000_clean) > 0 and len(df2_ac_000_clean) > 0:
    a = ks_2samp(df1_ac_000_clean, df2_ac_000_clean)
    print(f"\nKS test p-value: {a.pvalue}")
else:
    print("\nUnable to perform KS test: One or both arrays are empty after cleaning.")
```
